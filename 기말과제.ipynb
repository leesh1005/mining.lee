```python
google = google.drop(['Scraped Time'], axis=1)
```

```python
display(google)
```

```python
google.drop('Developer Id', axis=1)
google.drop('Developer Website', axis=1)
google.drop('Developer Email', axis=1)
google.drop('Privacy Policy', axis=1)
```

```python
google=google.dropna(subset=['App Name'],axis=0)
google=google.dropna(subset=['Currency'],axis=0)
google=google.dropna(subset=['Size'],axis=0)
google=google.dropna(subset=['Installs'],axis=0)
```

```python
google.isnull().sum()
```

```python
google.loc[google['Rating'] != google['Rating'], 'Rating'] = 0
google.loc[google['Rating Count'] != google['Rating Count'], 'Rating Count'] = 0
```

```python
google = google.astype({ 'Released' : 'string' })
```
